```python
import pandas as pd

print('bye')
```

```python
edu_df = pd.read_excel('/Users/bgrullon/challenges/ipc-county/data/Education.xlsx')
edu_df.rename(columns={
    'Federal Information Processing Standard (FIPS) Code': 'FIPS_Code'
}, inplace=True)
edu_df.drop(['State', 'Area name'], axis=1,inplace=True)
edu_df.head()
```

```python
unemp_df = pd.read_excel('/Users/bgrullon/challenges/ipc-county/data/Unemployment.xlsx')
unemp_df.head()
```

```python
pop_est_df = pd.read_excel('/Users/bgrullon/challenges/ipc-county/data/PopulationEstimates.xlsx')
pop_est_df.rename(columns={
    'FIPStxt': 'FIPS_Code'
}, inplace=True)
pop_est_df.drop(['State', 'Area_Name'], axis=1,inplace=True)
pop_est_df.head()
```

```python
m_df = pd.merge(unemp_df, pop_est_df, on='FIPS_Code')
m_df.head()
```

```python
df = pd.merge(m_df, edu_df, on='FIPS_Code')
df.head()
```
